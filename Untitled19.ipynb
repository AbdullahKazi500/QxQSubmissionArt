```python
%%manim BlochSphere
from manim import *
import numpy as np

class BlochSphere(Scene):
    def construct(self):
        # Define colors
        red = "#FF0000"
        green = "#00FF00"
        blue = "#0000FF"
        white = "#FFFFFF"
        black = "#000000"
        
        # Create sphere and axes
        sphere = Sphere(radius=2, stroke_color=white, stroke_width=1, fill_opacity=0)
        x_axis = Arrow(start=sphere.get_bottom(), end=sphere.get_top(), color=red, buff=0, stroke_width=2)
        y_axis = Arrow(start=sphere.get_left(), end=sphere.get_right(), color=green, buff=0, stroke_width=2)
        z_axis = Arrow(start=sphere.get_center(), end=sphere.get_north_pole(), color=blue, buff=0, stroke_width=2)
        
        # Add labels
        x_label = Text("x", color=red).next_to(x_axis.get_end(), RIGHT)
        y_label = Text("y", color=green).next_to(y_axis.get_end(), UP)
        z_label = Text("z", color=blue).next_to(z_axis.get_end(), RIGHT)
        ket_0 = TexMobject("\\left|0\\right\\rangle", color=white).scale(0.75).next_to(x_axis.get_start(), LEFT, buff=0.2)
        ket_1 = TexMobject("\\left|1\\right\\rangle", color=white).scale(0.75).next_to(x_axis.get_end(), RIGHT, buff=0.2)
        
        # Add background
        background = Rectangle(width=FRAME_WIDTH, height=FRAME_HEIGHT, fill_opacity=1, fill_color=black)
        self.add(background)
        
        # Add objects to scene
        self.add(sphere, x_axis, y_axis, z_axis, x_label, y_label, z_label, ket_0, ket_1)
        
        # Rotate Bloch sphere
        self.play(Rotate(sphere, angle=PI/2, axis=y_axis.get_unit_vector(), run_time=2))
        self.play(Rotate(sphere, angle=PI/2, axis=z_axis.get_unit_vector(), run_time=2))
```

```python
%%manim LinearEquationExample
from manim import *

class LinearEquationExample(Scene):
    def construct(self):
        eq1 = Tex(r"2x + y = 3")
        eq2 = Tex(r"x - y = 1")
        eq_group = VGroup(eq1, eq2).arrange(DOWN, buff=1)
        self.play(Create(eq_group))
        self.wait()
        self.play(eq1.animate.shift(LEFT), eq2.animate.shift(RIGHT))
        self.wait()
        self.play(eq_group.animate.scale(0.8).to_edge(UP))
        self.wait()
        matrix = Matrix([[2, 1], [1, -1]], h_buff=1.5)
        equals = Tex(r"\LARGE{=}")
        rhs = Matrix([[3], [1]], h_buff=1.5)
        matrix_group = VGroup(matrix, equals, rhs).arrange(RIGHT, buff=1.5)
        self.play(Transform(eq_group, matrix_group), run_time=2)
        self.wait()
        self.play(matrix[1][0].animate.set_color(RED))
        self.wait()
```

```python
cQuantumCircuit
from manim import *
from manim.mobject.qubit import Qubit
from manim.mobject.qubit_grid import QubitGrid
from manim.mobject.matrix import Matrix

class QuantumCircuit(Scene):
    def construct(self):
        # Create a qubit circle and rotate it
        qubit_circle = QubitCircle()
        qubit = Qubit("0")
        self.play(Create(qubit_circle), Create(qubit))
        self.play(Rotate(qubit_circle, PI/2), Rotate(qubit, PI/2))

        # Apply a single-qubit gate to the qubit
        pauli_x = Matrix([["0", "1"], ["1", "0"]])
        gate = QuantumGate(pauli_x)
        self.play(Create(gate))
        self.play(ApplyGate(gate, qubit))

        # Create a grid of qubits and apply a CNOT gate
        qubit_grid = QubitGrid(n_rows=2, n_cols=2)
        self.play(Create(qubit_grid))
        self.wait(1)
        self.play(Transform(qubit, qubit_grid[0,0]))
        self.play(Transform(qubit_grid[1,0], qubit_grid[0,1]))
        self.play(ApplyGate(CNOT(), (qubit, qubit_grid[1,0])))

        self.wait(1)
```
